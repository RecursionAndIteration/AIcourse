```python
import matplotlib.pyplot as plt
import numpy as np

x= np.random.rand(30)
y= np.random.rand(30)
plt.scatter(x,y,color='blue',marker="o", label="losowe punkty")
#plt.plot(x,y,color='blue',marker="o", label="losowe punkty")
plt.title("losowe punkty")
plt.legend()
plt.xlabel("os x")
plt.ylabel("os y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random

dice_rolls = np.array([(random.randint(1,7)) for _ in range (20)])
print(dice_rolls)

plt.hist(dice_rolls, bins = 12, color="blue",edgecolor="black")
plt.title("histogram rzutu kostka")
plt.xlabel("wartosc")
plt.ylabel("liczba wystapien")
plt.show()
```

```python
import matplotlib.pyplot as plt


reason = ["Luck","Skill","Concentrated power of will","Pleasure","Pain"]
percentage = [10,20,15,5,50]

plt.pie(percentage, labels=reason, autopct="%1.1f%%",colors=["red","green","blue","yellow","pink"])
plt.show()
```

```python
import matplotlib.pyplot as plt


label=["A","B","C","D"]
values=[10,15,50,25]

fig, axes = plt.subplots(1,2, figsize=(12,5))
axes[0].pie(values, labels=label, autopct="%1.1f%%", colors=["red","green","blue","yellow","pink"])
axes[0].set_title("wykres kolowy")

axes[1].bar(label,values)
axes[1].set_title("Wykres supkowy")
axes[1].set_xlabel("Kategorie")
axes[1].set_ylabel("Wartoci")
plt.tight_layout()
plt.show()
```
